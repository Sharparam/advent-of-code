```python
FILENAME = "input"
```

```python
from shapely import Point

with open(FILENAME) as file:
    points = [Point(*map(int, line.split(","))) for line in file]
```

```python
from itertools import combinations
from shapely import Polygon

def poly(a, b):
    x1, x2 = sorted((a.x, b.x))
    y1, y2 = sorted((a.y, b.y))
    return Polygon(((x1, y1), (x1, y2), (x2, y2), (x2, y1)))

boxes = [poly(a, b) for (a, b) in combinations(points, 2)]
```

```python
def area(box: Polygon) -> float:
    return (box.bounds[2] - box.bounds[0] + 1) * (box.bounds[3] - box.bounds[1] + 1)

int(max(area(box) for box in boxes))
```

```python
from shapely import Polygon

polygon = Polygon(points)

int(max(area(box) for box in boxes if polygon.contains(box)))
```
